```python
from bs4 import BeautifulSoup as BS
import re

def finddesc(X):
    list = ['RevPro', 'SQL', 'tableau', 'Saas','python','Stata', 'Excel', 'Word', 'Power Point', 'Microsoft Office', 'CRM', 'Oracle', 'Enterprise', 'finance', 'cloud', 'big data', 'accounting', 'forecasting', 'computer science', 'statistics', 'economics', 'model','modeling', 'marketing','sales','hospitality','healthcare','business','food','logistic','operation','entertainment','human resources', 'MBA', 'Bachelor', 'masters','master','phd', 'degree']
    match=['SAS','R','POC','SAP','BS','BA','MS','MA','BA/BS','BS/BA', 'Java']
    list = [k.lower() for k in list]
    match = [k.lower() for k in match]
    keyword = []
    exp=[]
    num=""
    htmlfile = urllib.urlopen(X) 
    soup = BS(htmlfile,'html5lib')
    content_list=soup.find('div',attrs={'class':"main_job"})
    pretty=content_list.prettify()
    pretty1=pretty.lower()
    
    for i in list:
        if(pretty1.find(i, 0) > 0):
            keyword.append(i)
            
    for i in match: 
        if re.search(r"\b{}\b".format(i), pretty1, re.IGNORECASE) is not None:
            keyword.append(i)
            
    if re.search(r'\b(years)\b', pretty1) is not None: 
        a=re.search(r'\b(years)\b', pretty1)
        a=a.start()
        num=pretty1[(a-10):a]
        if "one" in num:
            num="1"
        if "two" in num:
            num="2"
        if "three" in num:
            num="3"
        if "four" in num:
            num="4"
        if "five" in num:
            num="5"
        if "six" in num:
            num="6"
        if "seven" in num:
            num="7"
        if "eight" in num:
            num="8"
        if "nine" in num:
            num="9"
        if "ten" in num:
            num="10"
        num=re.sub("[^0-9_-]", "", num)
        if "-" in num:
            num = num.split('-', 1)[1]
            if "-" in num:
                num = num.split('-', 1)[1]
        exp.append(num)
    
    keyword=keyword+exp
    
    return keyword
    
frame2=pd.DataFrame(NewDf,columns=['company','date','title','locations','salary','actual_link','description'])
x = []

for index, row in NewDf.iterrows():
    url = row['actual_link']
    description = finddesc(url)
    x.append(description)
frame2['description'] = x
frame2
```

```python
list1 = ['RevPro', 'SQL', 'tableau', 'Saas','SAS','R','POC','SAP','python','Stata', 'Excel','Java', 'Word', 'Power Point', 'Microsoft Office', 'CRM', 'Oracle', 'Enterprise', 'finance', 'cloud', 'big data', 'accounting', 'forecasting', 'computer science', 'statistics', 'economics', 'model','modeling', 'marketing','sales','hospitality','entertainment','healthcare','business','food','logistic','operation','human resources', 'MBA', 'Bachelor', 'masters','master','phd','degree','BS','BA','MS','MA','BA/BS','BS/BA']
list1 = [k.lower() for k in list1]
final_df = pd.DataFrame(index = frame2.index, columns = list1)
final_df['description'] = frame2['description']
desc = final_df['description']
z = 0
for i in desc:
    for skill in i:
        final_df.loc[z, skill] = 1
    z += 1
final_df = final_df.fillna(0)
final_df
```

```python
pd.set_option('display.max_colwidth', -1) #expand full columns
final_df=final_df.drop(columns=['description'])
frames = [frame2, final_df]
result = pd.concat(frames, axis = 1)
result
```

```python
#result=result.drop(columns=['Bachelor'])
result['Bachelor'] = np.where((result['ba']==1) | (result['bs']==1) | (result['ba/bs']==1) | (result['bs/ba']==1) | (result['bachelor']==1),1,0)
result['Master'] = np.where((result['ma']==1) | (result['ms']==1) | (result['master']==1) | (result["masters"]==1) | (result['mba']==1),1,0)
result
```

```python
result1=result

result1.loc[result1.Bachelor==1, ['degree']] = 0
result1.loc[result1.Master==1, ['degree']] = 0
result1['Master_pref']=np.where((result['Bachelor']==1) & (result['Master']==1),1,0)
result1.loc[result1.Master_pref==1, ['Bachelor','Master']] = 0
result1=result1.drop(columns=['bs' , 'ba' , 'bs/ba' , 'ba/bs' , 'bachelor' , 'mba' , 'master' , 'masters' , 'ms' , 'ma' , 'phd'])
result1
```
